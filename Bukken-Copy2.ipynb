```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
bukken = pd.read_csv("bukken_train.csv")
bukken_test = pd.read_csv("bukken_test.csv")
bukken
```

```python
bukken.isnull().sum()
```

```python
bukken_kind = pd.get_dummies(bukken['frontal_road_kind'])
bukken_direction = pd.get_dummies(bukken['frontal_road_direction'])
bukken_shape = pd.get_dummies(bukken['land_shape'])

bukken2= pd.concat((bukken, bukken_kind, bukken_direction, bukken_shape), axis=1)
```

```python
bukken_test_kind = pd.get_dummies(bukken_test['frontal_road_kind'])
bukken_test_direction = pd.get_dummies(bukken_test['frontal_road_direction'])
bukken_test_shape = pd.get_dummies(bukken_test['land_shape'])

bukken2_test= pd.concat((bukken_test, bukken_test_kind, bukken_test_direction, bukken_test_shape), axis=1)
```

```python
bukken2_test
```

```python
bukken2.corr()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train = bukken2.loc[:,['minute_to_station','land_space','frontage_length','frontal_road_width','building_land_ratio','floor_land_ratio']].values

X_test = bukken2_test.loc[:,['minute_to_station','land_space','frontage_length','frontal_road_width','building_land_ratio','floor_land_ratio']].values
```

```python
y_train = bukken2.loc[:, ['price_per_tsubo']].values

y_test = bukken2_test.loc[:, ['price_per_tsubo']].values
```

```python
# 変数の標準化を実行
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train_std = ss.fit_transform(X_train)
X_test_std = ss.fit_transform(X_test)
```

```python
model = LinearRegression()

model.fit(X_train_std,y_train)
```

```python
model.intercept_
model.coef_
```

```python
print('R^2')
print('train: %.3f' % model.score(X_train_std, y_train))
print('test : %.3f' % model.score(X_test_std, y_test))
```

```python
from sklearn.metrics import mean_squared_error as mse
print('RMSE')
print('train: %.3f' % (mse(y_train, model.predict(X_train_std)) ** (1/2)))
print('test : %.3f' % (mse(y_test, model.predict(X_test_std)) ** (1/2)))
```
